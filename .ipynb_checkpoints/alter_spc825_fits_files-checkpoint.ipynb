# Import the necessary packages and define the path to the phaseb_data directory as well as the new directory you would like to save the altered files to. 

```python
from astropy.io import fits
import astropy.units as u
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'image.origin': 'lower'})
import os
from pathlib import Path
import misc

spcdir_old = Path('/groups/douglase/phaseb_data/spc_20181220'); print(spcdir_old)
spcdir = Path('/groups/douglase/webbpsf-data/CGI/optics/F730'); print(spcdir)
```

# Add pupil scales to the pupil files.

```python
pupil_fnames = ['pupil_SPC-20181220_1k.fits', 'pupil_SPC-20181220_1k_rotated.fits']
pupil_diam = 2.3633372*u.m
pupil_diam_pix = 1000*u.pixel

for fname in pupil_fnames:
    spc825pup = fits.open(spcdir_old/fname)
    spc825pup_data = spc825pup[0].data
    spc825pup.close()
    
    hdr = fits.Header()
    hdr['PUPLDIAM'] = pupil_diam.value
    hdr['PUPLSCAL'] = pupil_diam.value/pupil_diam_pix.value
    hdr['PIXELSCL'] = pupil_diam.value/pupil_diam_pix.value
    spc825pup_new = fits.PrimaryHDU(data=spc825pup_data, header=hdr)
    spc825pup_new.writeto(spcdir/fname, overwrite=True)
    print(spcdir/fname)
    print(pupil_diam/pupil_diam_pix, spc825pup_data.shape)
    plt.figure()
    plt.imshow(spc825pup_data)
```

# Add pupil scales to the apodizer files.

```python
spm_fnames = ['SPM_SPC-20181220_1000_rounded9_gray.fits', 'SPM_SPC-20181220_1000_rounded9_gray_rotated.fits']
pxscl = 1.7000e-5

for fname in spm_fnames:
    spc825spm = fits.open(spcdir_old/fname)
    spc825spm_data = spc825spm[0].data
    spc825spm.close()
    
    hdr = fits.Header()
    hdr['PIXSCALE'] = pxscl
    hdr['PUPLSCAL'] = pxscl*spc825spm_data.shape[0]
    hdr['PIXELSCL'] = pxscl
    spc825spm_new = fits.PrimaryHDU(data=spc825spm_data, header=hdr)
    spc825spm_new.writeto(spcdir/fname, overwrite=True)
    print(spcdir/fname)
    print(pxscl, spc825spm_data.shape)
    plt.figure()
    plt.imshow(spc825spm_data)
```

# Add pixel scales to the FPM files.

```python
lambda_m = 825e-9*u.m
D = 2.3633372*u.m

lamD = lambda_m/D
as_per_lamD = lamD * 180/np.pi * 3600

fpm_fnames = ['FPM_res50_SPC-20181220.fits', 'fpm_0.05lamdivD.fits']
fpm_sampling = [(lamD/50*u.radian).to(u.arcsecond), (0.05*lamD*u.radian).to(u.arcsecond)]

for i,fname in enumerate(fpm_fnames):
    spc825fpm = fits.open(spcdir_old/fname)
    spc825fpm_data = spc825fpm[0].data
    spc825fpm.close()
    
    hdr = fits.Header()
    hdr['PIXELSCL'] = fpm_sampling[i].value
    spc825fpm_new = fits.PrimaryHDU(data=spc825fpm_data, header=hdr)
    spc825fpm_new.writeto(spcdir/fname, overwrite=True)
    print(spcdir/fname)
    print(fpm_sampling[i], spc825fpm_data.shape)
    plt.figure()
    plt.imshow(spc825fpm_data)
```

# Add pupil scales to the lyot stop files.

```python
ls_fnames = ['LS_SPC-20181220_1k.fits', 'LS_SPC-20181220_1k_rotated.fits', 
             'lyotstop_0.5mag.fits', 'lyotstop_0.5mag_rotated.fits']
ls_sampling = [1.7000e-5*u.m, 1.7000e-5*u.m,
               1.7000e-5*1002/501*u.m, 1.7000e-5*1002/501*u.m]

for i,fname in enumerate(ls_fnames):
    spc825ls = fits.open(spcdir_old/fname)
    ls_data = spc825ls[0].data
    spc825ls.close()
    
    hdr = fits.Header()
    hdr['PUPLDIAM'] = ls_sampling[i].value*ls_data.shape[0]-2
    hdr['PUPLSCAL'] = ls_sampling[i].value
    hdr['PIXELSCL'] = ls_sampling[i].value
    ls_new = fits.PrimaryHDU(data=ls_data, header=hdr)
    ls_new.writeto(spcdir/fname, overwrite=True)
    print(str(spcdir/fname))
    print(ls_data.shape, ls_sampling[i])
    plt.figure()
    plt.imshow(ls_data)
```
